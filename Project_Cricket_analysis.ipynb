#**T20 Cricket World Cup Data Pre Processing:**

```python
import pandas as pd
import json
```

# Match result


```python
# Opening the file
with open('/content/Json files/t20_wc_match_results.json') as f :
  data = json.load(f)

# Converting data to dataframe using pandas
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
# Rename scorecard to match_id, sothat it will work as primary key
df_match.rename({'scorecard':'match_id'}, axis = 1, inplace =True)
df_match.head()
```

```python
# Saving the file
df_match.to_csv('dim_match_summary.csv')
```

```python
# creating match id dictionary to connect with other table

match_ids_dict ={}

for index,row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_ids_dict[key1]=row['match_id']
  match_ids_dict[key2]=row['match_id']

match_ids_dict
```

# Batting Summary

```python
with open('/content/Json files/t20_wc_batting_summary.json') as f:
  data = json.load(f)

  all_record =[]
# To get single list of all matches
  for rec in data:
    all_record.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_record)
df_batting.head()
```

```python
# Created new column to check the wether player out or not out from existing column dismissal
df_batting['out/not_out'] = df_batting['dismissal'].apply(lambda x : 'out' if len(x) >0 else 'not out')
df_batting.head(11)
```

```python
# Deleting dismissal column
df_batting.drop(columns =['dismissal'], inplace =True)
df_batting.head(11)
```

```python
df_batting.head(11)
```

```python
# Removing special symball present in batsmanName

df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))

df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)

df_batting.head(11)

```

```python
# Saving the file
df_batting.to_csv('fact_batting_summary.csv', index = False)
```

# Bowling Summary :

```python
with open('/content/Json files/t20_wc_bowling_summary.json') as f:
  data = json.load(f)

  all_record = []
  for rec in data:
    all_record.extend(rec['bowlingSummary'])

df_bowling = pd.DataFrame(all_record)
df_bowling.head(10)
```

```python
# Adding new table match_id taking refrance from match table

df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
# Saving to csv file
df_bowling.to_csv('fact_bowling_summary.csv',index = False)
```
